```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.spatial import Voronoi, voronoi_plot_2d
import scipy.ndimage as ndimage

optimo_weisz = [394.15476007, 224.64950489]
optimo_prox = [385.87786274, 234.05272394]
optimo_random = [357.62820918, 255.83589184]
optimo_random2 = [386.06721191, 233.851883  ]

df = pd.read_csv(os.path.join("data","points.csv"), sep=",").drop("peso", axis = 1)
df2 = pd.read_csv(os.path.join("data","peso_random.csv"), sep=",").drop("peso",axis = 1)
df3 = pd.read_csv(os.path.join("data","perturbacion_puntos.csv"), sep=",").drop("peso",axis = 1)
```

```python
#Coordenadas de recintos de salud ya existentes
points = np.array(df)
#Coordenadas con candidato a ptimo calculado con Alg. de Weiszfeld
points_optimo = np.append(points,[optimo_weisz],axis = 0)
#Construccin de diagrama de Voronoi de los recintos de salud ya existentes
vor = Voronoi(points) 
#[385.87774582, 234.05265303]
#Construccin de diagrama de Voronoi de los recintos de salud junto al candidato a ptimo
vor2 = Voronoi(points_optimo)

#Definiciones para graficar los diagramas de Voronoi
img_file = 'Region de Estudio.png'
img = np.flipud(plt.imread(img_file))
img = (img  * 255).astype(np.uint8)
fig = plt.figure(figsize=(20,20))
ax = fig.add_subplot(221)
ax2 = fig.add_subplot(222)

ax.imshow(ndimage.rotate(img, 360), origin='lower')
ax2.imshow(ndimage.rotate(img, 360), origin='lower')
voronoi_plot_2d(vor, point_size=10, ax=ax)
voronoi_plot_2d(vor2, point_size=10, ax=ax2)
ax.set_ylim(ymin=0,ymax = 560)
ax.set_xlim(xmin=0,xmax = 561)
ax2.set_xlim(xmin=0,xmax = 561)
ax2.set_ylim(ymin=0,ymax = 560)
ax.title.set_text('Sin optimo')
ax2.title.set_text('Con optimo')
plt.show()
```
